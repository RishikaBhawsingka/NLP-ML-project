```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2)
)
```

```python
train["clean_text"].isnull().sum()
```

```python
train["clean_text"] = train["clean_text"].fillna("")
validation["clean_text"] = validation["clean_text"].fillna("")
test["clean_text"] = test["clean_text"].fillna("")
```

```python
X_train = tfidf.fit_transform(train["clean_text"])
X_validation = tfidf.transform(validation["clean_text"])
X_test = tfidf.transform(test["clean_text"])  #text to number
```

```python
y_train = train["label"]
y_validation = validation["label"]
y_test = test["label"]
```

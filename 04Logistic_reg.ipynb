```python
import pandas as pd
import joblib

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)
```

```python
train = pd.read_csv(
    "/content/drive/MyDrive/HumanBehaviourSimulator/cleaned_data/train_clean.csv",
    keep_default_na=False
)

validation = pd.read_csv(
    "/content/drive/MyDrive/HumanBehaviourSimulator/cleaned_data/validation_clean.csv",
    keep_default_na=False
)

test = pd.read_csv(
    "/content/drive/MyDrive/HumanBehaviourSimulator/cleaned_data/test_clean.csv",
    keep_default_na=False
)
```

```python
tfidf = joblib.load(
    "/content/drive/MyDrive/HumanBehaviourSimulator/models/tfidf_vectorizer.pkl"
)
```

```python
#text to number
X_train = tfidf.transform(train["clean_text"])
X_validation = tfidf.transform(validation["clean_text"])
X_test = tfidf.transform(test["clean_text"])
```

```python
y_train = train["label"]
y_validation = validation["label"]
y_test = test["label"]
```

```python
lr_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)
```

```python
#training of model
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_validation)
```

```python
accuracy = accuracy_score(y_validation, y_pred)

precision = precision_score(
    y_validation,
    y_pred,
    average="weighted"
)

recall = recall_score(
    y_validation,
    y_pred,
    average="weighted"
)

f1 = f1_score(
    y_validation,
    y_pred,
    average="weighted"
)

print("Accuracy :", accuracy)
print("Precision:", precision)
print("Recall   :", recall)
print("F1 Score :", f1)
```

```python
print(classification_report(y_validation, y_pred))
```

```python
cm = confusion_matrix(y_validation, y_pred)

print(cm)
```

```python
joblib.dump(
    lr_model,
    "/content/drive/MyDrive/HumanBehaviourSimulator/models/logistic_regression.pkl"
)

print("✅ Logistic Regression model saved!")
```

```python
lr_model_balanced = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    random_state=42
)

lr_model_balanced.fit(X_train, y_train)


# Predictions on validation data
y_pred_balanced = lr_model_balanced.predict(X_validation)
```

```python
joblib.dump(
    lr_model_balanced,
    "/content/drive/MyDrive/HumanBehaviourSimulator/models/logistic_regression_balanced.pkl"
)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

accuracy = accuracy_score(y_validation, y_pred_balanced)

precision = precision_score(
    y_validation,
    y_pred_balanced,
    average="weighted",
    zero_division=0
)

recall = recall_score(
    y_validation,
    y_pred_balanced,
    average="weighted",
    zero_division=0
)

f1 = f1_score(
    y_validation,
    y_pred_balanced,
    average="weighted",
    zero_division=0
)

print("="*50)
print("📊 Logistic Regression (Balanced)")
print("="*50)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
```

```python
print(classification_report(
    y_validation,
    y_pred_balanced,
    zero_division=0
))
```

```python
cm = confusion_matrix(y_validation, y_pred_balanced)

print(cm)
```
